```python
import cv2
import matplotlib.pyplot as plt

def yolo_format(path):
    image =cv2.imread("sample_30k_virtualdataset_resized/"+path+".jpg")

    dh, dw, _ = image.shape
    with open("sample_30k_virtualdataset/"+path+".txt") as f:
        data = f.readlines()
        f.close()
    
    for dt in data:
        cls, x, y, w, h = map(float, dt.split())
        if cls !=78945:
            l = int((x - w / 2) * dw)
            r = int((x + w / 2) * dw)
            t = int((y - h / 2) * dh)
            b = int((y + h / 2) * dh)
    
            if l < 0:
                l = 0
            if r > dw - 1:
                r = dw - 1
            if t < 0:
                t = 0
            if b > dh - 1:
                b = dh - 1

            cv2.rectangle(image, (l, t), (r, b), (0, 0, 255), 2)
    return image
```

```python
def center_crop(img, dim):
    img =cv2.resize(img,(0,0) ,fx =0.9,fy =0.9)
    width, height = img.shape[1], img.shape[0]
    crop_width = dim[0] if dim[0]<img.shape[1] else img.shape[1]
    crop_height = dim[1] if dim[1]<img.shape[0] else img.shape[0]
    mid_x, mid_y = int(width/2), int(height/2)
    cw2, ch2 = int(crop_width/2), int(crop_height/2) 
    crop_img = img[mid_y-ch2:mid_y+ch2, mid_x-cw2:mid_x+cw2]
    return crop_img
```
